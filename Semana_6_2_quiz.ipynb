# Contexto:
Un permiso de construcción es un documento oficial de aprobación emitido por una agencia gubernamental que le permite a usted o a su contratista proceder con un proyecto de construcción o remodelación en su propiedad. Para más detalles, visite https://www.thespruce.com/what-is-a-building-permit-1398344. Cada ciudad o condado tiene su propia oficina relacionada con los edificios, que puede realizar múltiples funciones como la emisión de permisos, la inspección de edificios para hacer cumplir las medidas de seguridad, la modificación de las normas para adaptarse a las necesidades de la creciente población, etc. En el caso de la ciudad de San Francisco, la expedición de permisos corre a cargo de www.sfdbi.org/
